```python
wine_red = pd.read_csv('winequality-red.csv')
```

```python
display(wine_red.head())
```

```python
from sklearn.decomposition import PCA

X = wine_red.iloc[:,:12].values
pca = PCA(n_components=3)

X_pca = pca.fit_transform(X)
wine_red_new = pd.DataFrame(X_pca)
wine_red_new.columns=['f1','f2','f3']
display(wine_red_new.head())
```

```python
from matplotlib import pyplot as plt
%matplotlib inline
plt.scatter(wine_red_new['f1'],wine_red_new['f2'])
plt.show()
```
